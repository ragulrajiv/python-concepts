```python
# importing the libararies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x=np.arange(0,10,1)
y=2*x + 5

#changing the figure
fig=plt.figure(figsize=(10,5))


#ploting the datapoints
#ALPHA=VISIBILITY
plt.plot(x ,y,linewidth = 2.0,linestyle = "--", color = 'black', alpha = 0.7, marker= 'h',markerfacecolor = 'r')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['Line1'],loc='best')
plt.grid(True)
plt.show()
```

sub plot

```python
# importing the libararies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x=np.arange(0,10,1)
y1=2*x + 5
y2=3*x+10


#create two or more plots in fig
plt.subplot(1,2,1)               #gives the chart in horizontal or vetrtical(1,2-vertical   , 2,1  is horizontal)
plt.plot(x,y1,color='r')
plt.grid("True")
plt.title("y1-graph")

plt.subplot(1,2,2)
plt.plot(x,y2,color='b')
plt.title("y2-graph")
plt.grid("True")
plt.show()


#ploting the datapoints
#ALPHA=VISIBILITY
plt.plot(x,y1,color='r',marker='*')

plt.plot(x,y2)
plt.title("SINGLE GRAPH")
plt.grid("True")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(figsize=(20,20))

#1.PLOT
x=np.arange(0,10,1)
y=x*2+5
plt.subplot(4,4,1)
plt.plot(x,y,color='BLACK',marker='h',markerfacecolor='r',linestyle='--')

plt.title("LINE GRAPH")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.grid(True)
plt.legend(['*'],loc='best')


#2.BAR CHART
data={'red':10,
      'blue':20,
      'green':30,
      'yellow':40,
      'white':50}
names=list(data.keys())
values=list(data.values())
plt.subplot(4,4,2)
plt.bar(names,values,color=['gold','c','b','k'])
plt.title("BAR GRAPH")
plt.xlabel("COLOR")
plt.ylabel("VALUES")
plt.grid(True)
#plt.legend(['*'],loc='best')



#3.SCATTER CHART
#craeting data set
a=np.arange(10,90,10)
b=np.random.randint(1,9,8)
c=np.random.randint(1,9,8)
plt.subplot(4,4,3)

#creating a scatter plot

plt.scatter(a,b,color='m', marker='*')

plt.scatter(a,c, color='b')

plt.plot(a,b)
plt.plot(a,c)
plt.title("SCATTER GRAPH")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.grid(True)
plt.legend(['a','b','c'],loc='best')



#4.PIE CHART

#PREAPRING THE DATASET
animal=['dog','cat','wolf','lion']
count=[50,45,60,80]
plt.subplot(4,4,4)

#creating a pie chart
plt.pie(count,labels= animal)
plt.title("PIE CHART")


plt.legend(['*'],loc='best')




#customizing pie chart
#PREAPRING THE DATASET

animal=['dog','cat','wolf','lion']
count=[50,45,60,80]
plt.subplot(4,4,5)

#creating a pie chart
plt.pie(count,labels= animal,autopct='%1.1f%%',shadow='True',explode=(0,0.2,0,0),startangle= 90)
plt.title("PIE CHART DEMO")



#5.histogram graph
#PREAPRING THE DATASET
# preparaing the dataset for age
age=np.random.randint(1,100,20)
#creating the histogram
plt.subplot(4,4,6)
plt.hist(age, bins=[0,20,40,60,80,100])    #bins-5
plt.title("HISTOGRAM CHART")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.grid(True)


#6.DONUT GRAPH
color=['red','blue','white','green']
count=[50,45,60,80]
size_center=[10]
plt.subplot(4,4,7)

pie1=plt.pie(count,labels=color,radius=1.5,autopct='%1.1f%%')
#plt.pie(count,labels= color,radius=1.5,autopct='%1.1f%%',shadow='True',explode=(0,0.2,0,0),startangle= 90)
pie2=plt.pie(size_center,radius=1.0,colors='w')


#8.spilt garph

#defining the dataset
x=np.arange(0,10,1)
y1=2*x + 5
y2=3*x+10


#create two or more plots in fig
plt.subplot(4,4,8)               #gives the chart in horizontal or vetrtical(1,2-vertical   , 2,1  is horizontal)
plt.plot(x,y1,color='r',marker='*')
plt.title("y1-graph")
plt.xlabel("X-AXIS")
plt.ylabel("Y1-AXIS")
plt.grid(True)

plt.subplot(4,4,9)
plt.plot(x,y2,color='b',marker="h")
plt.title("y2-graph")
plt.xlabel("X-AXIS")
plt.ylabel("Y2-AXIS")
plt.grid(True)




#ploting the datapoints
#ALPHA=VISIBILITY
plt.subplot(4,4,10)
plt.plot(x,y1,color='r',marker='*')
plt.plot(x,y2,color='black',marker='p')
plt.title("SINGLE GRAPH")

plt.xlabel("X-AXIS")
plt.ylabel("Y1&2-AXIS")
plt.grid("True")





#adjust layout
plt.tight_layout()
plt.show()
```
